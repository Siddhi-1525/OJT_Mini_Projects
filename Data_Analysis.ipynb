```python
df = pd.read_csv('Students Social Media Addiction.csv')
df
```

```python
sns.countplot(x='Gender', data=df)
plt.title('Gender Distribution')
plt.show()
```

```python
print(df.columns)
```

```python
df.groupby('Gender')['Addicted_Score'].mean()
```

```python
df.sort_values(by='Addicted_Score', ascending=False).head()
```

```python
df['Gender'].value_counts()
```

```python
df.groupby('Age')['Addicted_Score'].mean().sort_values(ascending=False)
```

```python
print("Highest Addiction Score:\n", df[df['Addicted_Score'] == df['Addicted_Score'].max()])
print("\nLowest Addiction Score:\n", df[df['Addicted_Score'] == df['Addicted_Score'].min()])
```

```python
df.corr(numeric_only=True)
```

```python
plt.hist(df['Addicted_Score'], bins=10, color='skyblue', edgecolor='black')
plt.title('Addicted_Score Distribution')
plt.xlabel('Addicted_Score')
plt.ylabel('Number of Students')
plt.show()
```

```python
plt.scatter(df['Age'], df['Addicted_Score'], color='purple')
plt.title('Age vs Addicted_Score')
plt.xlabel('Age')
plt.ylabel('Addicted_Score')
plt.grid(True)
plt.show()
```
